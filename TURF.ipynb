```python
import pandas as pd
import numpy as np
import itertools

df = pd.read_csv('base_for_turf.csv', sep=';')
df = df.drop(df.columns[0], axis=1)

def calculate_combination_per_alternative(max_alternative=4):

  max_reach_list = []

  for i in range(1, max_alternative + 1):
    df_all_combs = pd.DataFrame({
        str(comb): df[list(comb)].sum(axis=1)
        for comb in itertools.combinations(df.columns, i)
    })

    reach_pct = (df_all_combs.astype(bool).sum(axis=0) / len(df)) * 100

    best_comb = reach_pct.idxmax()
    best_value = reach_pct.max()
    best_quantity = round(reach_pct.max() / 100 * len(df))
    max_reach_list.append({
        'Combination': best_comb,
        'Max reach (%)': best_value,
        'Quantity': best_quantity
    })

  result_df = pd.DataFrame(max_reach_list)
  result_df['Increase (%)'] = result_df['Max reach (%)'].diff().fillna('')
  return result_df

calculate_combination_per_alternative(7)
```
